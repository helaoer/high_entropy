```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
# Load data from Excel
df = pd.read_excel("./all_rp750_pred.xlsx")

# Use the second column as X and the first as Y
X = df.iloc[:, -1].values.reshape(-1, 1)  # Reshape for sklearn
Y = df.iloc[:, -2].values

rmse = np.sqrt(mean_squared_error(Y, X))
r2 = r2_score(Y, X)

print(f"original RMSE: {rmse}")
print(f"original R^2 score: {r2}")

# Train Linear Regression model
model = LinearRegression()
model.fit(X,Y)

# Predict on testing set
Y_pred = model.predict(X)

# Compute RMSE and R^2
rmse = np.sqrt(mean_squared_error(Y, Y_pred))
r2 = r2_score(Y, Y_pred)
with open('../label_scaler.pickle', 'rb') as handle:
    scaler = pickle.load(handle)
    mean = scaler.mean_[0]
    std = scaler.scale_[0]
print(mean,std)
Y_ = (Y-0.01-6.174468393277538)*std+mean
Y_pred_ = (Y_pred-0.01-6.174468393277538)*std+mean
rmse_ = np.sqrt(mean_squared_error(Y_, Y_pred_))

print(f"RMSE: {rmse}")
print(f"RMSE_original: {rmse_}")  
print(f"R^2 score: {r2}")

# Plotting true vs predicted values
plt.scatter(X, Y, color='black', label='All data')
plt.plot(X, Y_pred, color='blue', linewidth=3, label='Linear regression')
plt.xlabel('New descriptor')
plt.ylabel('RP750')
plt.title(f'RMSE: {rmse:.2f} RMSE_original: {rmse_:.4f} R2 :{r2:.2f}')
plt.legend()
plt.show()
```
